# 스포츠 뉴스

```python
sports = ['wbaseball', 'kbaseball', 'kfootball', 'wfootball', 'basketball', 'volleyball', 'golf', 'esports']
```

```python
link = []
for sport in sports:
    for num in range(3):
        n = num +1
        html = "https://sports.news.naver.com/"+sport+"/news/index.nhn?page="+str(n)
        URL = html
        print(URL)

        driver.get(URL)


        navernews = driver.find_elements_by_xpath("//li//div[@class='text']")

        for a in navernews:
            try:
                link.append (a.find_element_by_class_name("title").get_attribute('href'))
            except NoSuchElementException:
                pass
            except IndexError:
                pass
        
link  = list(filter(None, link))
```

```python
articles_sport=[]
articles_sport_title=[]

for i in range(len(link)):
    driver.get(link[i])
    print(link[i])
    
    content = driver.find_element_by_xpath("//div[@id='newsEndContents']")
    title = driver.find_element_by_xpath("//h4[@class='title']")
    article_sport=[]

    p = re.compile(".*다\..*\n" ,re.MULTILINE)
    a = p.findall(content.text)
        
    for k in range(len(a)):
        sub1 = re.sub('\s.*기자',' ',a[k]) 
        sub2 = re.sub('.디지털데일리.*\s','',sub1) 
        sub3 = re.sub('\n','',sub2)
        sub4 = re.sub('[\\\']',' ',sub3)
        sub5 = re.sub('\[.*?\]',' ',sub4)

        article_sport.append(sub5)
        
    title = re.sub('[\\\']',' ',title.text)
        
    articles_sport.append("".join(article_sport))
    articles_sport_title.append(title)
```

```python
import pandas as pd
df_sport = pd.DataFrame(columns=['Title','Content'])
df_sport['Title'] = articles_sport_title
df_sport['Content'] = articles_sport
```

```python
df_sport["Category"][0:60] = "해외야구"
df_sport["Category"][60:120] = "국내야구"
df_sport["Category"][120:180] = "국내축구"
df_sport["Category"][180:240] = "해외축구"
df_sport["Category"][240:300] = "농구"
df_sport["Category"][300:360] = "배구"
df_sport["Category"][360:420] = "골프"
df_sport["Category"][420:480] = "e스포츠"
```

```python
df_sport.to_csv('/home/ngh/바탕화면/sport.csv')
```

# gensim doc2vec

```python
import mecab
from konlpy.tag import Okt
from konlpy.tag import Kkma
from konlpy.tag import Komoran
from konlpy.tag import Hannanum
```

```python
def komoran_noun(doc):
    return Komoran().nouns(doc)
def komoran_morphs(doc):
    return Komoran().morphs(doc)

def mecab_noun(doc):
    return mecab.MeCab().nouns(doc)
def mecab_morphs(doc):
    return mecab.MeCab().morphs(doc)

def twitter_noun(doc):
    return Okt().nouns(doc)
def twitter_morphs(doc):
    return Okt().morphs(doc)
```

```python
df_sum['content_morphs'] = df_sum['Content'].apply(mecab_noun)
```

```python
df_sum2 = df_sum
df_sum2["Index"] = df_sum.index
```

```python
for i in range(len(df_sum2["Index"])):
    df_sum2["Index"][i] =  [str(df_sum2["Index"][i])+"번"]
```

```python
df_sum2
```

```python
from gensim.models import Doc2Vec
from gensim.models import doc2vec
```

```python
doc_vectorizer = Doc2Vec(
    window=5,        # distance between the predicted word and context words
    size=50,        # vector size
    alpha=0.025,     # learning-rate
    seed=1234,
    min_alpha=0.025, # min learning-rate
    hs = 1,          # hierarchical softmax / default 0
    negative = 5,   # negative sampling / default 5
    dbow_words = 0
)
```

```python
from collections import namedtuple
TaggedDocument = namedtuple('TaggedDocument', 'words tags')
tagged_train_docs = [TaggedDocument(d,c) for d,c in df_sum2[['content_morphs','Index']].values]
print(len(tagged_train_docs))
```

```python
tagged_train_docs[0:10]
```

```python
doc_vectorizer.build_vocab(tagged_train_docs)
```

```python
from time import time
start = time()
for epoch in range(10):
    doc_vectorizer.train(tagged_train_docs, total_examples=doc_vectorizer.corpus_count, epochs=5)
    doc_vectorizer.alpha -= 0.002 # decrease the learning rate
    doc_vectorizer.min_alpha = doc_vectorizer.alpha # fix the learning rate, no decay
end = time()
print("During Time: {}".format(end-start))
```

```python
doc_vectorizer.save('/home/ngh/바탕화면/doc2vec_nouns.model')
```
